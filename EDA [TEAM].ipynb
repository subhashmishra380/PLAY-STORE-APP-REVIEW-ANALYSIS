# Now we will have to remove all the unwanted characters to convert data type to numeric 

````python
def remove_it(num):
  """This function takes a string and replace the following characters if present, '+', ',' ,'$', 'M', 'k', 'NaN'"""
  if '+' in num:
    num = num.replace('+','')
  if ',' in num:
    num = num.replace(',','')
  if '$' in num:
    num = num.replace('$','')
  if 'M' in num:
    num = str(int(float(num.replace('M',''))*1000000))
  if 'k' in num:
    num = str(int(float(num.replace('k',''))*1000))
  if 'NaN' in num:
    num = '0'
  else:
    pass
  return num
````

````python
#Changing the 'Reviews' column values into valid numeric values
df1['Reviews'] = pd.to_numeric(df1['Reviews'])

#Changing the 'Size' column values into valid numeric values
df1['Size'] = df1['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)
df1['Size'] = pd.to_numeric(df1['Size'].map(lambda x: remove_it(x)))

#Changing the 'Installs' column values into valid numeric values
df1['Installs'] = pd.to_numeric(df1['Installs'].map(lambda x: remove_it(x)))

#Changing the 'Price' column values into valid numeric values
df1['Price'] = pd.to_numeric (df1['Price'].map(lambda x:remove_it(x)))
df1.info()
````
